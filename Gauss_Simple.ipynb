```python
import math

a,b,b1,mat,x=[],[],[],[],[]

def inicializar_Matriz(fil,col):
  
    mat= [None] * fil
    
    if(fil>0):
        for i in range(fil):
            mat[i] = [None] * col
            
    return mat

    
def copiar_matriz(a):
    
    mat=inicializar_Matriz(len(a),0);
    i,j=0,0
    
    while i<len(a):
        while j<len(a[i]):
            mat[i].append(a[i][j])
            j+=1
        j=0
        i+=1
    
    return mat

def llenar_Matriz(fil,col):
    
    global a
    global b
    
    a,b=inicializar_Matriz(fil,col),inicializar_Matriz(fil,0)
    
    i,j=0,0
    strg=""
    
    while i<len(a):
        while j<(len(a[i])):
            strg+="a["+str(i)+"]["+str(j)+"] = "
            a[i][j]=float(input(strg))
            j+=1
            strg=""
            
        b[i]=float(input("b["+str(i)+"]= "))
        
        j=0
        i+=1
    print(" ")

def pivoteo_Parcial(k):
    
    i,j,big,dummy=k,0,0,0
    p=k
    big=abs(mat[k][k])
    
    while i<len(mat):
        dummy=abs(mat[i][k])
        if dummy>big:
            big=dummy
            p=i
        i+=1
       
    j=k
    if(p!=k):
    
        while j<len(mat):
            dummy=mat[p][j]
            mat[p][j]=mat[k][j]
            mat[k][j]=dummy 
            j+=1
            
        dummy=b1[p]
        b1[p]=b1[k]
        b1[k]=dummy

def eliminacion_hacia_adelante(a,b):

    global b1,mat
    j,x,z,factor,mat,b1=0,0,0,0,copiar_matriz(a),b[:]

    while z<len(mat):
        
        pivoteo_Parcial(z)
        x=z+1
        while x<len(mat):
            factor=mat[x][z]/mat[z][z]
            while j<len(mat):
                mat[x][j]=mat[x][j]-factor*mat[z][j]
                j+=1
            j=0
            b1[x]=b1[x]-factor*b1[z]
            x+=1
        z+=1
        
def sustitucion_hacia_atras(mat):
    
    global x
    x=inicializar_Matriz(len(mat),0)
    
    i=len(mat)-1
    n=len(mat)-1
    
    while i>=0:
        aux=b1[i]
        j=i+1
        while j<=n:
            aux=aux-mat[i][j]*x[j]
            j+=1
        x[i]=aux/mat[i][i]
        i-=1 
    
    print(" ")
    i=0
    while i<len(x):
        print("X"+str(i+1)+"=",x[i])
        i+=1


llenar_Matriz(3,3)
eliminacion_hacia_adelante(a,b)
sustitucion_hacia_atras(mat)
```
