```python
import math
import matplotlib.pyplot as pl
%matplotlib inline

x,y=[],[]
xo,n=0,0

def insertarPuntos():
     global xo,n
     i,n=0,int(input("Numero de datos: "))
    
     while i<n:
           x.append(int(input("X"+str(i)+" ="))) 
           i+=1
        
def f(x):
    y=math.log(x)
    return y
  
def interpolacionLagrange(x,n,xo):
    
    i,j,lx,fo=0,0,1,0
    
    while i<n:
        lx=1
        while j<n:
            if(i!=j):
                lx*=(xo-x[j])/(x[i]-x[j])
            j+=1
        j=0
        fo+=lx*f(x[i])
        i+=1
    return fo
        
def graficar():
    
    xp=[1,4,6]
    xcd,xlin,fx,y=[],[],[],[]

    for i in range(1,50):
        fx.append(f(i))
        xcd.append(interpolacionLagrange(xp,3,i))
        xlin.append(interpolacionLagrange(xp,2,i))
        y.append(i)
     
    pl.title("Interpolacion de Lagrange")
    pl.ylabel('Y')
    pl.xlabel('X')
    
    pl.plot(fx,y,color="black",label = "f(x)=ln(x)",linewidth=2.5, linestyle="-")
    pl.plot(xlin,y,color="green",label = "Polinomio de primer grado",linewidth=2, linestyle="-")
    pl.plot(xcd,y,color="red",label = "Polinomio de segundo grado",linewidth=2, linestyle="-")
    pl.minorticks_on()
    pl.grid(True)
    pl.axis([0,10,0,10])
    pl.legend()
    pl.show()  
    
graficar()
```
