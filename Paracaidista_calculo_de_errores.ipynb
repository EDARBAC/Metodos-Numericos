```python
from tabulate import tabulate
import math
import matplotlib.pyplot as pl

g=9.8
m=68.1
c=12.5

eTlr=0.5*math.pow(10,-2)

vl=[]
tp=[]

vll=[]
tpp=[]

tabla=[]

eV=[]
eL=[]
eA=[]

def met_1():
    for to in range(215):
        
        v=((g*m)/c)*(1-math.exp(-(c/m)*to))
        vl.append(v)
    
def met_2():
    
    vo=0
    t=0
    ti=2
    
    vll.append(0.0)
    
    for i in range(214): 
            
        v=vo+(g-(c/m)*vo)*(ti-t)
        vll.append(v)
        t=ti
        ti=t+2
        vo=v
        
def error_ver():
    
    cont1=0
    cont2=0
    
    while cont1<len(vl):
        
        eV.append(abs(vl[cont1]-vll[cont2]))
        #print(eV[cont2])
        tp.append(cont1)
        cont1=cont1+2
        cont2=cont2+1
       
        
    
def error_por_ver():
    
    cont=1
    eL.append(0.0)
    
    while cont<len(eV):
        eL.append((eV[cont]/vll[cont])*100)
    
        cont=cont+1
    
def error_rel_por_apro():
    
    c1=0
    c2=0
    
    while (c1+1)<len(vll):
        tpp.append(c2)
        eA.append((((vll[c1+1]-vll[c1])*100)/vll[c1+1]))
        c1=c1+1
        c2=c1+1
        
def errorTolerado():
    
    t=0
    to=0
    
    print (len(vl), len(vll), len(tp), len(eV), len(eL), len(eA))
    
    while(eA[t]>eTlr  and to<len(vl)):
        tabla.append([tp[t],vl[to], vll[t], eV[t], eL[t], eA[t]])
        t+=1 
        to+=2
   
    print (tabulate(tabla, headers=[ "tiempo(s)","Velocidad(m/s)","Velocidad1(m/s)","Error Verdadero", "Error % Verdadero", 
                                    "Error Relativo % Aproximado"]))
           
def graf():
    
    pl.ylabel(' ')
    pl.xlabel('Tiempo (s)')
   
    pl.plot(tp,eV,"ro")
    pl.plot(tp,eL,color="green",linewidth=2.5, linestyle="-")
    pl.plot(tpp,eA,color="pink",linewidth=2.5, linestyle="-")
    pl.grid(True)
    pl.axis([-10,50,0,100])
    pl.show() 
    
met_1()
met_2()
error_ver()
error_por_ver()
error_rel_por_apro()
errorTolerado()
graf()
```
