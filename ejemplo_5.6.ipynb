```python
import math
import matplotlib.pyplot as pl
from tabulate import tabulate


rX1, rX2=[],[]
Et,Et1=[],[]
Ea,Ea1=[],[]
I1,I2=[],[]

tabla=[]
tabla1=[]

Es=0.5

vV=1

def f(x):
    
    y=(x**10)-1
    
    return y

def encontrarRaiz(Xl,Xu,ban):
    
    Xr=0.0
    Ett=0.6
    
    vAct=1
    vAnt=0
    
    c1,c2,i=0,0,0
    
    while Ett>Es:
        
        
        if(ban==True):
            Xr=((Xl+Xu)/2)
            rX1.append(Xr)
            Ett=abs((vV-Xr)/vV)*100
            Et.append(Ett)
            I1.append(c1+1)
            
            vAct=Xr
            Ea.append(abs(((vAct-vAnt)/vAct)*100))
            vAnt=vAct
            
            tabla.append([I1[c1],Xl,Xu,Xr,Ett,Ea[c1]])
            
            c1+=1
            
        else:
            Xr=Xu-(f(Xu)*(Xl-Xu)/(f(Xl)-f(Xu)))
            rX2.append(Xr)
            Ett=abs((vV-Xr)/vV)*100
            Et1.append(Ett)
            I2.append(c2+1)
            
            vAct=Xr
            Ea1.append(abs(((vAct-vAnt)/vAct)*100))
            vAnt=vAct
            
            tabla1.append([I2[c2],Xl,Xu,Xr,Ett,Ea1[c2]])
            
            c2+=1
        
        if(f(Xl)*f(Xr))<0:
            Xu=Xr
        
        else:
            if(f(Xl)*f(Xr))>0:
                Xl=Xr
            else:
                break
                
        i+=1
    
    print(len(tabla))       
    return Xr

def graf():
    
    encontrarRaiz(0,1.3,True)
    encontrarRaiz(0,1.3,False)
    
    pl.ylabel('f(X)')
    pl.xlabel('X')
   
    pl.plot(I1,rX1,color="green",linewidth=2.5, linestyle="-")
   
    pl.axis([0,15,0,6])
    pl.grid(True)
    pl.show()


graf()
   
print(tabulate(tabla,headers=["Iteracion","Xl","Xu","Xr","Error relativo porcentual Verdadero","Error Aproximado"]))
print("#############################################################################################")
print(tabulate(tabla1,headers=["Iteracion","Xl","Xu","Xr","Error relativo porcentual Verdadero","Error Aproximado"]))

    

```
