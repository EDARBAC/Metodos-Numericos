```python
import math
from tabulate import tabulate

vV=math.exp(-5)

vApx=[]
vApx1=[]

eV=[]
eV1=[]

eL=[]
eL1=[]

eA=[]
eA1=[]

def g(n,x):
    tab=[]
    c1=0
    c2=0
    
    g=0
    x=abs(x)
    
    
    while c1<n:
    
     if c1%2==0:
        g+=((x**c1)/math.factorial(c1))   
        
     else:
        g-=((x**c1)/math.factorial(c1))
        
     c1+=1
     vApx.append(g)
    
    while (c2+1)<len(vApx):
        
        #Error Verdadero
        eV.append(abs(vV-vApx[c2]))
        #Error relativo porcentual verdadero
        eL.append((eV[c2]/vApx[c2])*100)
        #Error Relativo porcentual Aproximado
        eA.append((((vApx[c2+1]-vApx[c2])/vApx[c2+1])*100))
        
        tab.append([vApx[c2],eV[c2],eL[c2],eA[c2]])
        c2+=1
        
    print(tabulate(tab, headers=["Valor Aprox", "Error Ver", "Error rel % v",
                                 "Error rel % A"],tablefmt="pipe"))
    print("")
    #return g

def g1(n,x):
    
    tab=[]
    c1=0
    c2=0
    
    x=abs(x)
    
    g1=0
    
    while c1<n:
        g1+=((x**c1)/math.factorial(c1))
        vApx1.append((1/g1))
        c1+=1
        
    while (c2+1)<len(vApx1):
        
        #Error Verdadero
        eV1.append(abs(vV-vApx1[c2]))
        #Error relativo porcentual verdadero
        eL1.append((eV1[c2]/vApx1[c2])*100)
        #Error Relativo porcentual Aproximado
        eA1.append(abs(((vApx1[c2+1]-vApx1[c2])/vApx1[c2+1])*100))
        #tabular
        tab.append([vApx1[c2],eV1[c2],eL1[c2],eA1[c2]])
        c2+=1
        
    print(tabulate(tab, headers=["Valor Aprox", "Error Ver", "Error rel % v",
                                 "Error rel % A"],tablefmt="pipe"))
    
    return (1/g1)

g(20,-5)
g1(20,-5)
```
