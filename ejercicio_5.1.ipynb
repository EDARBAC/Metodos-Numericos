```python
import math
import matplotlib.pyplot as pl
from tabulate import tabulate

Et=[]
Ea=[]

fo=[]
xo=[]

Xro=[]

tabla=[]

vVo=6.4051
Xr_my=0

def f(x):
    
    y=-0.5*x**2 + 2.5*x + 4.5
    
    return y

def met_biseccion(Xl,Xu):
    
    Xr=0.0
    
    vAct=0
    vAnt=0
    
    c1=0
    i=0
    
    global tabla
    
    while c1<3:
    
        Xr=((Xl+Xu)/2)
        Ett=abs((vVo-Xr)/vVo)*100
        Et.append(Ett)
        Xro.append(Xr)
        
        if(c1>0):
            Ea.append(abs((((Xro[c1]-Xro[c1-1])/Xro[c1])*100)))
            tabla.append([Xl,Xu,Xr,Ett,Ea[c1-1]])
        else:
            tabla.append([Xl,Xu,Xr,Ett,""])
            
            
            
        if(f(Xl)*f(Xr))<0:
            Xu=Xr
        
        else:
            if(f(Xl)*f(Xr))>0:
                Xl=Xr
            else:
                break
                
        c1+=1
    
    return Xr

def encontrarRaiz_form_cua():
    
    a=1
    b=-11
    c=4.5
    
    y=(-2.5+math.sqrt((b**2)-4*a*c))/(2*a)
    yo=(-2.5-math.sqrt((b**2) - 4*a*c))/(2*a)
    
    print(chr(27)+"[1;47m"+" Raiz 1 : %f    Raiz 2 : %f"%(y,yo))
   

def graf():
    
    print("")
    print("Raiz mas grande por metodo de biseccion:",vVo)
    
    x=-3
    
    while x<20:
        fo.append(f(x))
        xo.append(x)
        x+=1
    
    pl.ylabel('f(X)')
    pl.xlabel('X')
   
    pl.plot(xo,fo,color="green",linewidth=2.5, linestyle="-")
    pl.plot(6.405125,0,'ro',color="pink",label="Raiz mas grande",linewidth=3)
   
    pl.axis([-5,15,-4,8])
    pl.grid(True)
    pl.legend()
    pl.show()


encontrarRaiz_form_cua()
met_biseccion(5,10)

print(chr(27)+"[0m")
headers=["Xl","Xu","Xr","Et","Ea","Raices"]

print(tabulate(tabla,headers))
graf()
    

```
