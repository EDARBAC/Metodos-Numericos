```python
import math
import matplotlib.pyplot as pl
from tabulate import tabulate

fo=[]
xo=[]

def f(x):
    
    y=math.log(x**2)-0.7
    
    return y

def encontrarRaiz(Xl,Xu,ban):
    
    Ea=[]
    tabla=[]
    Xro=[]
    Es=0
    
    Xr=0.0
    Et=1
    Eapx=1
    
    c1=0
    
    while c1<3:
    
        if(ban==True):  
            Es=0.1
            
            Xr=((Xl+Xu)/2)
            Xro.append(Xr)
        
            if(c1>0):
                Eapx=(abs((((Xro[c1]-Xro[c1-1])/Xro[c1])*100)))
                Ea.append(Eapx)
                tabla.append([c1+1,Xl,Xu,Xr,Eapx])
            
            else:
                tabla.append([c1+1,Xl,Xu,Xr,""])
                
        else:
            Es=0.002
            Xr=Xu-(f(Xu)*(Xl-Xu)/(f(Xl)-f(Xu)))
            Xro.append(Xr)
            
            if(c1>0):
                Eapx=(abs((((Xro[c1]-Xro[c1-1])/Xro[c1])*100)))
                Ea.append(Eapx)
                tabla.append([c1+1,Xl,Xu,Xr,Eapx])
            
            else:
                tabla.append([c1+1,Xl,Xu,Xr,""])
        
        
        if(f(Xl)*f(Xr))<0:
            Xu=Xr
        
        else:
            if(f(Xl)*f(Xr))>0:
                Xl=Xr
            else:
                break
                
        c1+=1
        
    headers=["Iteracion","Xl","Xu","Xr","Ea"]
    print(tabulate(tabla,headers))
    
def graf():
    
    print("")
    print("Valor exacto: ",1.4190)
    print("")
    
    x=-30
    n=0
    
    while n<500:
        fo.append(f(x))
        xo.append(x)
        x+=0.1
        n+=1
    
    pl.ylabel('f(X)')
    pl.xlabel('X')
   
    pl.plot(xo,fo,color="green",linewidth=2.5, linestyle="-")
   
    pl.axis([-8,8,-100,100])
    pl.grid(True)
    pl.show()


graf()
#metodo de biseccion
encontrarRaiz(0.5,2,True)
print("")
#metodo de falsa posicion
encontrarRaiz(0.5,2,False)
```
