```python
import math
import matplotlib.pyplot as pl
from tabulate import tabulate

vV=0.56714329


def f(x):
    
    y=math.exp(-x)-x
    
    return y


def metodo_biseccion(Xl,Xu):
    
    tabla=[]
    rX=[]
    rX.append(0)
    Et=[]
    Et.append(100)
    Ea=[]
    Ea.append("")
    Ever=[]
    Ever.append("")
    I1=[]
    I1.append(0)
    
    Xr=0.0

    vAct=1
    vAnt=0
    
    c1=0
    
    while c1<20:

        Xr=((Xl+Xu)/2)
            
        rX.append(Xr)
        Et.append(abs((vV-Xr)/vV)*100)
        Ever.append(vV-Xr)
        I1.append(c1+1)
            
        vAct=Xr
        Ea.append(abs(((vAct-vAnt)/vAct)*100))
        vAnt=vAct
            
        tabla.append([I1[c1],rX[c1],Et[c1],Ea[c1],Ever[c1]])
        
        if(f(Xl)*f(Xr))<0:
            Xu=Xr
        
        else:
            if(f(Xl)*f(Xr))>0:
                Xl=Xr
            else:
                break
            
        c1+=1
        
    print(tabulate(tabla,headers=["Iteracion","Xr","Et","Ea","Ev"]))
    pl.plot(I1,Et,"ro",color="red",label="Et metodo biseccion",linewidth=2.5, linestyle="-")

def falsa_posicion(Xl,Xu):
    
    tabla=[]
    rX=[]
    rX.append(0)
    Et=[]
    Et.append(100)
    Ea=[]
    Ea.append("")
    Ever=[]
    Ever.append("")
    I1=[]
    I1.append(0)
    
    Xr=0.0
    
    vAct=1
    vAnt=0
    
    c1=0
    
    while (c1<20):
        
        Xr=Xu-(f(Xu)*(Xl-Xu)/(f(Xl)-f(Xu)))
    
        rX.append(Xr)
        Ever.append(vV-Xr)
        Et.append(abs((vV-Xr)/vV)*100)
        I1.append(c1+1)
    
        vAct=Xr
        Ea.append(abs(((vAct-vAnt)/vAct)*100))
        vAnt=vAct
    
        tabla.append([I1[c1],rX[c1],Et[c1],Ea[c1],Ever[c1]])
        
        if(f(Xl)*f(Xr))<0:
            Xu=Xr
        
        else:
            if(f(Xl)*f(Xr))>0:
                Xl=Xr
            else:
                break
        
        c1+=1
        
    print(tabulate(tabla,headers=["Iteracion","Xr","Et","Ea","Ev"]))
    pl.plot(I1,Et,"ro",color="cyan",label="Et metodo de falsa posicion",linewidth=2.5, linestyle="-")
    
def metodo_graf():
    
    n=0
    xo=0
    
    x=[]
    fo=[]
    
    while(n<100):
    
        fo.append(f(xo))
        x.append(xo)
        
        xo+=1
        n+=1
        
    #pl.plot(x,fo,color="black",label="f(x)=e^-x -x",linewidth=2.5, linestyle="-")
    
    
def metodo_punto_fijo():
    
    tabla=[]
    rX=[] 
    rX.append(0)
    Et=[]
    Et.append(100)
    Ea=[]
    Ea.append("")
    Ever=[]
    Ever.append("")
    I1=[]
    I1.append(0)
    
    c1=0
    Xr=0.0
    
    vAct=1
    vAnt=0
    
    
    while(c1<20):
        
        Xr=f(Xr)+Xr
        
        rX.append(Xr)
        Et.append(abs((vV-Xr)/vV)*100)
        Ever.append(vV-Xr)
        I1.append(c1+1)
        
        vAct=Xr
        Ea.append(abs(((vAct-vAnt)/vAct)*100))
        vAnt=vAct
        
        tabla.append([I1[c1],rX[c1],Et[c1],Ea[c1],Ever[c1]])
        
        c1+=1
    
    print(tabulate(tabla,headers=["Iteracion","Xr","Et","Ea","Ev"]))
    pl.plot(I1,Et,"ro",color="green",label="Et punto fijo",linewidth=2.5, linestyle="-")
    
def met_Newton_Raphson():
    
    tabla=[]
    rX=[] 
    rX.append(0)
    Et=[] 
    Et.append(100)
    Ea=[] 
    Ea.append("")
    Ever=[] 
    Ever.append("")
    I1=[] 
    I1.append(0)
    
    c1=0
    Xr=0.0
    
    vAct=1
    vAnt=0
    
    while(c1<20):
        
        Xr=Xr-(f(Xr)/(-math.exp(-Xr)-1))
        
        rX.append(Xr)
        Et.append(abs((vV-Xr)/vV)*100)
        Ever.append(vV-Xr)
        I1.append(c1+1)
        
        vAct=Xr
        Ea.append(abs(((vAct-vAnt)/vAct)*100))
        vAnt=vAct
        
        tabla.append([I1[c1],rX[c1],Et[c1],Ea[c1],Ever[c1]])
        
        c1+=1
    
    print(tabulate(tabla,headers=["Iteracion","Xr","Et","Ea","Ev"]))
    pl.plot(I1,Et,"ro",color="blue",label="Et met. Newton Raphson",linewidth=2.5, linestyle="-")
    
def met_secante():
    
    tabla=[]
    rX=[] 
    rX.append(0)
    Et=[]
    Et.append(100)
    Ea=[]
    Ea.append("")
    Ever=[]
    Ever.append("")
    I1=[]
    I1.append(0)
    
    Xr=0.0
    Xo=0
    X1=1
    
    c1=0
    
    vAct=1
    vAnt=0
    
    while(c1<7):
        
        Xr=X1-(f(X1)*(Xo-X1))/(f(Xo)-f(X1))
        Xo=X1
        X1=Xr
        
        rX.append(Xr)
        Et.append(abs((vV-Xr)/vV)*100)
        Ever.append(vV-Xr)
        I1.append(c1+1)
        
        vAct=Xr
        Ea.append(abs(((vAct-vAnt)/vAct)*100))
        vAnt=vAct
        
        tabla.append([I1[c1],rX[c1],Et[c1],Ea[c1],Ever[c1]])
        
        c1+=1
    
    print(tabulate(tabla,headers=["Iteracion","Xr","Et","Ea","Ev"]))
    pl.plot(I1,Et,"ro",color="yellow",label="Et met. Secante",linewidth=2.5, linestyle="-")
    
def met_secante_mod():
    
    tabla=[]
    rX=[] 
    rX.append(0)
    Et=[]
    Et.append(100)
    Ea=[]
    Ea.append("")
    Ever=[]
    Ever.append("")
    I1=[]
    I1.append(0)
    
    Xr=0.0
    Xo=1
    
    c1=0
    
    vAct=1
    vAnt=0
    

    while(c1<20):
        
        Xr=Xo-(f(Xo)*0.01)/(f(Xo+0.01)-f(Xo))
        Xo=Xr
        
        rX.append(Xr)
        Et.append(abs((vV-Xr)/vV)*100)
        Ever.append(vV-Xr)
        I1.append(c1+1)
        
        vAct=Xr
        Ea.append(abs(((vAct-vAnt)/vAct)*100))
        vAnt=vAct
        
        tabla.append([I1[c1],rX[c1],Et[c1],Ea[c1],Ever[c1]])
        
        c1+=1
    
    print(tabulate(tabla,headers=["Iteracion","Xr","Et","Ea","Ev"]))
    pl.plot(I1,Et,"ro",color="m",label="Et met. Secante Modificado",linewidth=2.5, linestyle="-")
    
def graficar():
    
    pl.axis([0,20,0.0000001,10])
    pl.legend()
    pl.show()

    
metodo_graf()
print("Metodo de Biseccion")
metodo_biseccion(0,1)
print("")
print("Metodo de Falsa Posicion")
falsa_posicion(0,1)
print("")
print("Metodo de punto fijo")
metodo_punto_fijo()
print("")
print("Metodo de Newton Raphson")
met_Newton_Raphson()
print("")
print("Metodo de la Secante")
met_secante()
print("")
print("Metodo de la Secante Modificada")
met_secante_mod()
graficar()
```
